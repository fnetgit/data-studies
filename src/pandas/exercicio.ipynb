```python
import pandas as pd

df = pd.read_csv("../../data/supermarket_sales.csv")

df.head()
```

1. Qual o faturamento total por filial?

```python
df.groupby("City")[["Total", "gross income"]].sum()
```

2. Qual a percentual de participação na receita de cada tipo de filial?

```python
(df.groupby("Product line")["Total"].sum() / df.groupby("Product line")["Total"].sum().sum()).sort_values() * 100
```

3. Como está distribuída o tipo de produto consumido por gênero?

```python
df.groupby(["Product line", "Gender"])[["Total"]].sum().pivot_table(index="Product line", columns="Gender")
```

4. Qual foi o faturamento por mês?

```python
df["Date"] = pd.to_datetime(df["Date"])
df["Month"] = df["Date"].apply(lambda x: x.month)
df["Year"] = df["Date"].apply(lambda x: x.year)
df.groupby("Month")["Total"].sum()
```

5. Qual a média de avaliação por cada filial em janeiro de 2019?

```python
df[(df["Year"] == 2019) & (df["Month"] == 1)]["Rating"].mean()
```

6. Como se distribui o gasto por tipo de consumidor em cada filial?

```python
df.groupby(["Customer type", "City"])["Total"].sum()
```
